```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("ML471_S1_Datafile_Practice.csv")
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df.set_index("Date", inplace=True)
df_monthly = df["Close"].resample("MS").mean()

result = seasonal_decompose(df_monthly, model="additive")

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 12), sharex=True)
result.observed.plot(ax=ax1)
ax1.set_title("Additive Decomposition: Observed")
result.trend.plot(ax=ax2)
ax2.set_title("Additive Decomposition: Trend")
result.seasonal.plot(ax=ax3)
ax3.set_title("Additive Decomposition: Seasonal")
result.resid.plot(ax=ax4)
ax4.set_title("Additive Decomposition: Residual")

plt.tight_layout()
plt.show()
```

```python

result = seasonal_decompose(df_monthly, model="multiplicative")

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 12), sharex=True)
result.observed.plot(ax=ax1)
ax1.set_title("Multiplicative Decomposition: Observed")
result.trend.plot(ax=ax2)
ax2.set_title("Multiplicative Decomposition: Trend")
result.seasonal.plot(ax=ax3)
ax3.set_title("Multiplicative Decomposition: Seasonal")
result.resid.plot(ax=ax4)
ax4.set_title("Multiplicative Decomposition: Residual")

plt.tight_layout()
plt.show()
```
