```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ML471_S1_Datafile_Concept.csv")
df["DATE"] = pd.to_datetime(df["DATE"])
df.head()
```

```python
df["Consumption_Diff"] = df["Consumption"].diff()
plt.plot(df["DATE"], df["Consumption_Diff"])
plt.title("Differenced Power Consumption")
plt.xlabel("Year")
plt.ylabel("Differenced Power Consumption")
plt.show()
```
