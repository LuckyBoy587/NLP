```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ML471_S1_Datafile_Concept.csv")
df["DATE"] = pd.to_datetime(df["DATE"])
df.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
# Set DATE as index and set frequency to monthly
df_decomp = df.set_index('DATE')
df_decomp.index.freq = 'MS'

# Perform seasonal decomposition
result = seasonal_decompose(df_decomp['Consumption'], model='additive')

# Plot the decomposition
plt.rcParams['figure.figsize'] = (12, 8)
result.plot()
plt.show()
```
